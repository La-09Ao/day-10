```python
import pandas as pd

# 1) Create a Series with custom labels
products = pd.Series([700, 150, 300],
                     index=["Laptop", "Mouse", "Keyboard"])

# 2) Access price of 'Laptop' (label-based indexing)
laptop_price = products["Laptop"]

# 3) Slice first two products (positional indexing)
first_two = products.iloc[0:2]

# 4) Print results
print("Full Series:\n", products)
print("\nPrice of Laptop:", laptop_price)
print("\nFirst two products:\n", first_two)
```

```python
import pandas as pd

# 1) Create the Series
grades = pd.Series([85, None, 92, 45, None, 78, 55])

# 2) Identify missing values
print("Missing values:\n", grades.isnull())

# 3) Fill missing values with 0
filled_grades = grades.fillna(0)

# 4) Boolean masking: scores > 60
passed = filled_grades[filled_grades > 60]

# 5) Print results
print("\nOriginal Series:\n", grades)
print("\nFilled Series:\n", filled_grades)
print("\nScores > 60:\n", passed)
```

```python
import pandas as pd

# 1) Create the Series
usernames = pd.Series([' Alice ', 'bOB', ' Charlie_Data ', 'daisy'])

# 2) Clean the text
cleaned = usernames.str.strip().str.lower()

# 3) Check which names contain 'a'
contains_a = cleaned.str.contains('a')

# 4) Print results
print("Cleaned Usernames:\n", cleaned)
print("\nContains letter 'a':\n", contains_a)
```
